#  **Employee Retention Prediction**

<b>Employee retention prediction is a **proactive business strategy** that uses **data analytics and AI** to stop valuable employees from leaving before they even submit a resignation. Instead of reacting to an employee's exit, companies analyze workforce data to identify **"flight risks"** early.  

This allows organizations to address issues such as **burnout, lack of growth opportunities, or job dissatisfaction**, helping them retain talented employees and maintain productivity.<b>


-----------

# **How It Works**:

# **1. Predictive Modeling**
<b>Companies use **Machine Learning models** like **Random Forest** and **XGBoost** to analyze employee data and predict the risk of attrition.<b>

# **2. Stay Interview**
<b>Managers conduct **stay interviews** to understand employee concerns, expectations, and satisfaction levels.<b>

# **3. Massive Cost Savings**
<b>Reducing employee turnover helps companies save significant costs related to **recruitment, onboarding, and training**.<b>

# **4. Targeted Training**
<b>Providing **continuous learning and development programs** improves employee engagement, skill development, and long-term retention.<b>

>**__1. Univariate Analysis__: Here, we'll focus on one feature at a time to understand its distribution and range.**
>

# **4.2 Univariate Analysis**

<div style="border-radius:10px; padding: 15px; background-color: #facfc8; font-size:115%; text-align:left">

**We undertake univariate analysis on the dataset's features, based on their datatype:**

**1. For __continuous data__:** **We plot histograms to gain insight into the distribution of each feature. This allows us to understand the central tendency, spread, and shape of the dataset's distribution.**

    
**2. For __categorical data__:** **Bar plots are utilized to visualize the frequency of each category. This provides a clear representation of the prominence of each category within the respective feature.**

**By employing these visualization techniques, we're better positioned to understand the individual characteristics of each feature in the dataset.**

# **4.2.1 Numerical Variables Univariate Analysis**


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
import numpy as np

# -----------------------------
# General style
# -----------------------------
sns.set_theme(style="whitegrid", context="notebook")
plt.rcParams["figure.dpi"] = 120
plt.rcParams["axes.titleweight"] = "bold"
plt.rcParams["axes.labelweight"] = "semibold"

# -----------------------------
# Only continuous numerical features
# -----------------------------
continuous_features = ["city_development_index", "training_hours"]
n_num = len(continuous_features)
ncols_num = 2
nrows_num = math.ceil(n_num / ncols_num)

fig, axes = plt.subplots(
    nrows=nrows_num,
    ncols=ncols_num,
    figsize=(18, 7 * nrows_num)
)

axes = np.array(axes).flatten()

num_colors = ["#d62828", "#e76f51", "#f28482", "#bc4749", "#c1121f", "#ff758f"]

for i, col in enumerate(continuous_features):
    data = train_df[col].dropna()

    sns.histplot(
        data=data,
        bins=25,
        kde=True,
        ax=axes[i],
        color=num_colors[i % len(num_colors)],
        edgecolor="white",
        linewidth=0.8,
        alpha=0.85
    )

    # -----------------------------
    # Titles and labels
    # -----------------------------
    axes[i].set_title(f"{col} Distribution", fontsize=16, pad=12)
    axes[i].set_xlabel(col, fontsize=12, labelpad=10)
    axes[i].set_ylabel("Count", fontsize=12, labelpad=10)
    axes[i].grid(axis="y", linestyle="--", alpha=0.35)

    # -----------------------------
    # Give enough top space so labels and stats never touch
    # -----------------------------
    ymin, ymax = axes[i].get_ylim()
    axes[i].set_ylim(ymin, ymax * 1.35)

    # -----------------------------
    # Statistics box - fixed safely at top-left
    # -----------------------------
    stats_text = (
        f"Mean : {data.mean():.2f}\n"
        f"Median : {data.median():.2f}\n"
        f"Std Dev : {data.std():.2f}"
    )

    axes[i].text(
        0.04,
        0.93,
        stats_text,
        transform=axes[i].transAxes,
        ha="left",
        va="top",
        fontsize=10,
        fontweight="bold",
        color="white",
        bbox=dict(
            boxstyle="round,pad=0.4",
            facecolor="#a4161a",
            edgecolor="white",
            alpha=0.98
        )
    )

    # -----------------------------
    # Value labels for each plot
    # -----------------------------
    patches = axes[i].patches

    if len(patches) > 0:
        if len(patches) > 15:
            label_fontsize = 8
            y_offset = 3
        else:
            label_fontsize = 10
            y_offset = 5

        y_top = axes[i].get_ylim()[1]

        for p in patches:
            height = p.get_height()
            if height > 0:
                label_y = min(height + (y_top * 0.01), y_top * 0.95)

                axes[i].annotate(
                    f"{int(height)}",
                    (p.get_x() + p.get_width() / 2, label_y),
                    ha="center",
                    va="bottom",
                    fontsize=label_fontsize,
                    fontweight="bold",
                    color="#3d0000",
                    clip_on=True,
                    bbox=dict(
                        boxstyle="round,pad=0.15",
                        facecolor="white",
                        edgecolor="none",
                        alpha=0.75
                    )
                )

    # -----------------------------
    # Clean ticks
    # -----------------------------
    axes[i].tick_params(axis="x", labelsize=10)
    axes[i].tick_params(axis="y", labelsize=10)

# Hide unused subplot if any
for j in range(i + 1, len(axes)):
    axes[j].axis("off")

# -----------------------------
# Main title
# -----------------------------
fig.suptitle(
    "Continuous Variables Univariate Analysis",
    fontsize=22,
    fontweight="bold",
    color="#7f0000",
    y=0.98
)

plt.tight_layout(rect=[0, 0, 1, 0.94], pad=3.2, w_pad=3.2, h_pad=3.2)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
import numpy as np

# -----------------------------
# General style
# -----------------------------
sns.set_theme(style="whitegrid", context="notebook")
plt.rcParams["figure.dpi"] = 120
plt.rcParams["axes.titleweight"] = "bold"
plt.rcParams["axes.labelweight"] = "semibold"
n_cat = len(cat_cols_train)
ncols_cat = 2
nrows_cat = math.ceil(n_cat / ncols_cat)

fig, axes = plt.subplots(
    nrows=nrows_cat,
    ncols=ncols_cat,
    figsize=(20, 5.8 * nrows_cat)
)
axes = np.array(axes).flatten()

cat_colors = ["#ff6b6b", "#f4978e", "#e5383b", "#ff8fa3", "#c9184a", "#d90429"]

for i, col in enumerate(cat_cols_train):
    value_counts = train_df[col].fillna("Missing").value_counts()

    # Show top 10 only for high-cardinality columns
    if value_counts.shape[0] > 10:
        value_counts = value_counts.head(10)

    # Plot
    sns.barplot(
        x=value_counts.values,
        y=value_counts.index,
        ax=axes[i],
        color=cat_colors[i % len(cat_colors)]
    )

    # Titles and labels
    axes[i].set_title(f"Distribution of {col}", fontsize=15, pad=12)
    axes[i].set_xlabel("Count", fontsize=12, labelpad=10)
    axes[i].set_ylabel(col, fontsize=12, labelpad=10)
    axes[i].grid(axis="x", linestyle="--", alpha=0.35)

    # Add enough right-side space so labels stay outside bars but inside plot
    max_count = value_counts.max()
    axes[i].set_xlim(0, max_count * 1.16)

    # Bold visible labels outside bars, but never beyond plot boundary
    right_limit = axes[i].get_xlim()[1]
    for p in axes[i].patches:
        w = p.get_width()
        if w > 0:
            # default: just outside bar
            x_text = w + max_count * 0.012

            # keep label inside the axes with a safe right margin
            x_text = min(x_text, right_limit - max_count * 0.035)

            axes[i].annotate(
                f"{int(w)}",
                (x_text, p.get_y() + p.get_height() / 2),
                ha="left",
                va="center",
                fontsize=9.8,
                fontweight="bold",
                color="black",
                clip_on=True,
                bbox=dict(
                    boxstyle="round,pad=0.15",
                    facecolor="white",
                    edgecolor="none",
                    alpha=0.9
                )
            )

    # Clean ticks
    axes[i].tick_params(axis="x", labelsize=10)
    axes[i].tick_params(axis="y", labelsize=10)

# Hide extra subplots
for j in range(i + 1, len(axes)):
    axes[j].axis("off")

fig.suptitle(
    "Categorical Variables Univariate Analysis",
    fontsize=22,
    fontweight="bold",
    color="#7f0000",
    y=0.98
)

plt.tight_layout(rect=[0, 0, 1, 0.96], pad=3.0, w_pad=3.0, h_pad=3.0)
plt.show()
```

# **4.3.2 Categorical Features vs Target**


<div style="border-radius:10px; padding: 15px; background-color: #facfc8; font-size:115%; text-align:left">

**It display __100% stacked bar plots__ for each categorical feature illustrating the proportion of each category across the two target classes, complemented by the exact counts and percentages on the bars.**

```python
from matplotlib.colors import ListedColormap
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
import textwrap

# Remove 'target' from categorical_features
categorical_features = [feature for feature in cat_cols_train if feature != 'target']

# Dynamic subplot layout
n_cat = len(categorical_features)
ncols = 4
nrows = math.ceil(n_cat / ncols)

fig, ax = plt.subplots(nrows=nrows, ncols=ncols, figsize=(24, 6.8 * nrows))
ax = np.array(ax).reshape(nrows, ncols)

# Define colormap
cmp = ListedColormap(['#ff826e', 'red'])

for i, col in enumerate(categorical_features):
    x, y = i // ncols, i % ncols
    curr_ax = ax[x, y]

    # Fill missing values
    temp_col = train_df[col].fillna('Missing')

    # Counts and proportions
    cross_tab = pd.crosstab(index=temp_col, columns=train_df['target'])
    cross_tab_prop = pd.crosstab(index=temp_col, columns=train_df['target'], normalize='index')

    # Keep only top categories if too many
    if len(cross_tab) > 6:
        top_categories = cross_tab.sum(axis=1).sort_values(ascending=False).head(6).index
        cross_tab = cross_tab.loc[top_categories]
        cross_tab_prop = cross_tab_prop.loc[top_categories]

    # Sort categories by target=1 proportion if available
    if 1 in cross_tab_prop.columns:
        ordered_index = cross_tab_prop.sort_values(by=1, ascending=True).index
        cross_tab = cross_tab.loc[ordered_index]
        cross_tab_prop = cross_tab_prop.loc[ordered_index]

    # Wrapped labels for long category names
    wrapped_index = [
        "\n".join(textwrap.wrap(str(idx), width=12)) if len(str(idx)) > 12 else str(idx)
        for idx in cross_tab_prop.index
    ]

    # Plot
    cross_tab_prop.plot(
        kind='bar',
        ax=curr_ax,
        stacked=True,
        width=0.72,
        colormap=cmp,
        legend=False
    )

    # Title and axes
    curr_ax.set_title(f'{col}', fontsize=15, fontweight='bold', pad=14)
    curr_ax.set_xlabel('')
    curr_ax.set_ylabel('Proportion', fontsize=10, fontweight='bold')
    curr_ax.set_ylim(0, 1.22)
    curr_ax.set_xticklabels(wrapped_index, rotation=0, ha='center', fontsize=8.5)
    curr_ax.tick_params(axis='y', labelsize=8.5)
    curr_ax.grid(axis='y', linestyle='--', alpha=0.28)

    # Labels: cleaner logic
    for idx, cat in enumerate(cross_tab.index):
        cumulative = 0

        for cls in cross_tab_prop.columns:
            proportion = cross_tab_prop.loc[cat, cls]
            count = cross_tab.loc[cat, cls]

            if proportion == 0:
                cumulative += proportion
                continue

            # Big enough segment -> place inside
            if proportion >= 0.14:
                curr_ax.text(
                    idx,
                    cumulative + proportion / 2,
                    f'{count}\n({proportion*100:.1f}%)',
                    ha='center',
                    va='center',
                    fontsize=8,
                    fontweight='bold',
                    color='black',
                    bbox=dict(
                        boxstyle='round,pad=0.18',
                        facecolor='white',
                        edgecolor='none',
                        alpha=0.82
                    ),
                    clip_on=True,
                    zorder=5
                )

            # Medium segment -> compact inside label
            elif proportion >= 0.08:
                curr_ax.text(
                    idx,
                    cumulative + proportion / 2,
                    f'{count}\n{proportion*100:.1f}%',
                    ha='center',
                    va='center',
                    fontsize=7,
                    fontweight='bold',
                    color='black',
                    bbox=dict(
                        boxstyle='round,pad=0.12',
                        facecolor='white',
                        edgecolor='none',
                        alpha=0.8
                    ),
                    clip_on=True,
                    zorder=5
                )

            # Small segment -> label above the stack to avoid overlap
            else:
                curr_ax.text(
                    idx,
                    min(cumulative + proportion + 0.035, 1.08),
                    f'{count}',
                    ha='center',
                    va='bottom',
                    fontsize=7,
                    fontweight='bold',
                    color='black',
                    bbox=dict(
                        boxstyle='round,pad=0.10',
                        facecolor='white',
                        edgecolor='none',
                        alpha=0.85
                    ),
                    clip_on=True,
                    zorder=5
                )

            cumulative += proportion

    # Clean legend placement - moved higher so it never overlaps the subplot title/text
    curr_ax.legend(
        title='target',
        labels=[str(c) for c in cross_tab_prop.columns],
        loc='upper center',
        bbox_to_anchor=(0.5, 1.24),
        ncol=len(cross_tab_prop.columns),
        fontsize=7.5,
        title_fontsize=8,
        frameon=True,
        borderaxespad=0.6,
        handlelength=1.2,
        columnspacing=0.8
    )

# Hide unused subplots
for j in range(i + 1, nrows * ncols):
    x, y = j // ncols, j % ncols
    ax[x, y].axis('off')

plt.suptitle('Categorical Features vs Target Stacked Barplots', fontsize=22, fontweight='bold', y=1.03)
plt.tight_layout(rect=[0, 0, 1, 0.94], w_pad=2.2, h_pad=3.0)
plt.show()
```

# **5.3 OUTLIER TREATMENT**

```python
import numpy as np
import pandas as pd

# ==============================
# 1. OUTLIER DETECTION USING IQR
# ==============================

# Calculate quartiles
Q1 = df[continuous_features].quantile(0.25)
Q3 = df[continuous_features].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Detect outliers
outliers_mask = (df[continuous_features] < lower_bound) | (df[continuous_features] > upper_bound)

# Count outliers
outliers_count = outliers_mask.sum()

# Calculate percentage
outliers_percentage = (outliers_count / len(df)) * 100

# Create summary table
outlier_summary = pd.DataFrame({
    "Lower Bound": lower_bound,
    "Upper Bound": upper_bound,
    "Outlier Count": outliers_count,
    "Outlier Percentage (%)": outliers_percentage
})

print("Outlier Summary:")
display(outlier_summary)


# ===================================
# 2. OUTLIER TREATMENT (IQR CAPPING)
# ===================================

def cap_outliers_iqr(dataframe, columns):
    """
    Caps outliers using IQR method.
    Values below lower bound are replaced with lower bound,
    and values above upper bound are replaced with upper bound.
    """

    df_out = dataframe.copy()

    for col in columns:

        Q1 = df_out[col].quantile(0.25)
        Q3 = df_out[col].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        df_out[col] = np.clip(df_out[col], lower_bound, upper_bound)

    return df_out


# Apply outlier treatment
X = cap_outliers_iqr(X, continuous_features)

print("Outlier treatment completed successfully.")
```

# **5.5 Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
def get_scaler(scale_type="standard"):
    """
    Returns the scaler object based on the selected scaling method.

    Parameters
    ----------
    scale_type : str
        Options:
        - 'standard'
        - 'minmax'
        - 'robust'
        - None

    Returns
    -------
    scaler object or None
    """
    if scale_type is None:
        return None
    elif scale_type == "standard":
        return StandardScaler()
    elif scale_type == "minmax":
        return MinMaxScaler()
    elif scale_type == "robust":
        return RobustScaler()
    else:
        raise ValueError("scale_type must be one of: None, 'standard', 'minmax', 'robust'")


def model_needs_scaling(model_name):
    """
    Returns True only for scale-sensitive models.

    In this project:
    - Logistic Regression -> scaling required
    - Random Forest -> scaling not required
    - XGBoost -> scaling not required
    - LightGBM -> scaling not required
    """
    model_name = model_name.lower().strip()

    scaling_required_models = {
        "logistic_regression",
        "logistic regression"
    }

    return model_name in scaling_required_models


def scaling_reason(model_name):
    """
    Returns proper explanation for whether scaling is needed.
    """
    model_name = model_name.lower().strip()

    if model_needs_scaling(model_name):
        return (
            f"Scaling applied for {model_name} because this model is sensitive to "
            f"feature magnitude and performs better when numerical features are on a comparable scale."
        )
    elif model_name in {"random_forest", "random forest", "xgboost", "lightgbm"}:
        return (
            f"Scaling skipped for {model_name} because it is a tree-based model. "
            f"Tree-based algorithms split data using threshold rules and are generally scale-insensitive."
        )
    else:
        return (
            f"Scaling skipped for {model_name} by default because it is not listed as a scale-sensitive model."
        )


def apply_feature_scaling(X_train, X_test, numerical_columns, model_name, scale_type="standard"):
    """
    Applies feature scaling only when the selected model requires it.

    Parameters
    ----------
    X_train : pd.DataFrame
    X_test : pd.DataFrame
    numerical_columns : list
        Numerical columns to scale
    model_name : str
        Name of the model
    scale_type : str
        'standard', 'minmax', 'robust', or None

    Returns
    -------
    X_train_scaled, X_test_scaled, scaler
    """
    X_train_scaled = X_train.copy()
    X_test_scaled = X_test.copy()
    scaler = None

    if model_needs_scaling(model_name):
        scaler = get_scaler(scale_type)

        X_train_scaled[numerical_columns] = scaler.fit_transform(X_train_scaled[numerical_columns])
        X_test_scaled[numerical_columns] = scaler.transform(X_test_scaled[numerical_columns])

    print(scaling_reason(model_name))

    return X_train_scaled, X_test_scaled, scaler


def prepare_scaled_data_for_models(X_train, X_test, numerical_columns, model_scale_map):
    """
    Prepares scaled/unscaled datasets for multiple models using one common function.

    Parameters
    ----------
    X_train : pd.DataFrame
    X_test : pd.DataFrame
    numerical_columns : list
        List of numerical columns
    model_scale_map : dict
        Example:
        {
            "logistic_regression": "standard",
            "random_forest": None,
            "xgboost": None,
            "lightgbm": None
        }

    Returns
    -------
    prepared_data : dict
        Dictionary containing scaled train/test data and scaler for each model
    """
    prepared_data = {}

    for model_name, scale_type in model_scale_map.items():
        X_train_model, X_test_model, scaler = apply_feature_scaling(
            X_train=X_train,
            X_test=X_test,
            numerical_columns=numerical_columns,
            model_name=model_name,
            scale_type=scale_type
        )

        prepared_data[model_name] = {
            "X_train": X_train_model,
            "X_test": X_test_model,
            "scaler": scaler,
            "scale_type": scale_type
        }

    return prepared_data
```

# **Step 6 | Feature Engineering**

```python
def apply_feature_engineering(dataframe):
    df_fe = dataframe.copy()

    # 1. Experience cleaning
    if "experience" in df_fe.columns:
        df_fe["experience_cleaned"] = df_fe["experience"].replace({
            "<1": "0",
            ">20": "21"
        })
        df_fe["experience_cleaned"] = pd.to_numeric(df_fe["experience_cleaned"], errors="coerce")

    # 2. Last new job cleaning
    if "last_new_job" in df_fe.columns:
        df_fe["last_new_job_cleaned"] = df_fe["last_new_job"].replace({
            "never": "0",
            ">4": "5"
        })
        df_fe["last_new_job_cleaned"] = pd.to_numeric(df_fe["last_new_job_cleaned"], errors="coerce")

    # 3. Training hours group
    if "training_hours" in df_fe.columns:
        df_fe["training_hours_group"] = pd.cut(
            df_fe["training_hours"],
            bins=[-np.inf, 20, 50, 100, np.inf],
            labels=["Low", "Medium", "High", "Very High"]
        )

    # 4. City development group
    if "city_development_index" in df_fe.columns:
        df_fe["city_development_group"] = pd.cut(
            df_fe["city_development_index"],
            bins=[-np.inf, 0.60, 0.75, 0.90, np.inf],
            labels=["Low", "Medium", "High", "Very High"]
        )

    # 5. Missing flags
    for col in ["gender", "major_discipline", "company_size", "company_type"]:
        if col in df_fe.columns:
            df_fe[f"{col}_missing_flag"] = df_fe[col].isnull().astype(int)

    # 6. Experience group
    if "experience_cleaned" in df_fe.columns:
        df_fe["experience_group"] = pd.cut(
            df_fe["experience_cleaned"],
            bins=[-np.inf, 2, 5, 10, 20, np.inf],
            labels=["Fresher", "Junior", "Mid", "Senior", "Expert"]
        )

    # 7. Combined education + experience
    if "education_level" in df_fe.columns and "experience_group" in df_fe.columns:
        df_fe["education_experience_combo"] = (
            df_fe["education_level"].astype(str) + "_" + df_fe["experience_group"].astype(str)
        )

    return df_fe
```

```python
import numpy as np

def transform_skewed_features(train_df, test_df, skew_threshold=1.0):
    train_df = train_df.copy()
    test_df = test_df.copy()

    # Automatically get ONLY numeric columns to avoid '50-99' string errors
    numeric_cols = train_df.select_dtypes(include=[np.number]).columns
    skewed_cols = []

    for col in numeric_cols:
        # Calculate skewness only on numeric data
        skewness = train_df[col].skew()

        # Check if skewness is high and data is non-negative for log transform
        if skewness > skew_threshold:
            if (train_df[col] >= 0).all():
                skewed_cols.append(col)

    # Apply log transformation
    for col in skewed_cols:
        train_df[col] = np.log1p(train_df[col])
        # Ensure test_df also gets the transform (handle potential negative values in test)
        test_df[col] = np.log1p(test_df[col].clip(lower=0))

    return train_df, test_df, skewed_cols

# Call the function without needing the problematic 'num_cols' list
X_train, X_test, transformed_cols = transform_skewed_features(
    X_train,
    X_test,
    skew_threshold=1.0
)

print(f"Transformation Complete.")
print(f"Skewed columns transformed: {transformed_cols}")
```

  <p><b>SMOTE – HANDLING CLASS IMBALANCE</b></p>

  <p><b>SMOTE stands for Synthetic Minority Oversampling Technique. It is a data preprocessing technique used to handle class imbalance in machine learning datasets.</b></p>

  <p><b>In many real-world datasets, the target variable is imbalanced, meaning one class (for example, <span style="font-weight:bold;">target = 0</span>) has significantly more observations than the other class (such as <span style="font-weight:bold;">target = 1</span>). This imbalance can cause machine learning models to become biased toward the majority class and perform poorly on the minority class.</b></p>

  <p><b>SMOTE solves this problem by creating synthetic (artificial) samples for the minority class instead of simply duplicating existing data. It generates new data points by interpolating between existing minority class samples, which helps the model learn better patterns.</b></p>

  <p><b>HOW SMOTE WORKS</b></p>

  <p><b>SMOTE selects a minority class data point and finds its nearest neighbors. It then creates a new synthetic point along the line between the original point and its neighbor. This process is repeated until the minority class is balanced with the majority class.</b></p>

  <p><b>WHY SMOTE IS IMPORTANT</b></p>

  <p><b>Using SMOTE improves model performance by ensuring that both classes are equally represented. It helps the model avoid bias toward the majority class and improves metrics such as recall, precision, and F1-score for the minority class.</b></p>

<p><b>The dataset is first divided into input features (X) and the target variable (y). Then, the data is split into training and testing sets. SMOTE is applied only on the training data to avoid data leakage.</b></p>

  <p><b>The <span style="font-weight:bold;">fit_resample()</span> function generates synthetic samples for the minority class and balances the dataset. After this step, the model can be trained on a more balanced dataset.</b></p>

  <p><b>FINAL CONCLUSION</b></p>

  <p><b>SMOTE is an effective technique to handle class imbalance by generating synthetic data for the minority class. It improves model fairness and performance, especially in classification problems where one class is underrepresented.</b></p>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.metrics import accuracy_score, classification_report
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

# =========================================================
# FIX ONLY THE LIGHTGBM FEATURE-NAME WARNING
# =========================================================
warnings.filterwarnings(
    "ignore",
    message="X does not have valid feature names, but LGBMClassifier was fitted with feature names"
)

# =========================================================
# 1. SPLIT DATA
# =========================================================
# Assumes X and y already exist
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

# Keep as DataFrame / Series
X_train = pd.DataFrame(X_train).copy()
X_test = pd.DataFrame(X_test).copy()
y_train = pd.Series(y_train).copy()
y_test = pd.Series(y_test).copy()

# =========================================================
# 2. DETECT COLUMN TYPES
# =========================================================
numeric_features = X_train.select_dtypes(
    include=["int64", "float64", "int32", "float32"]
).columns.tolist()

categorical_features = X_train.select_dtypes(
    include=["object", "category", "bool"]
).columns.tolist()

print("Numeric columns:", numeric_features)
print("Categorical columns:", categorical_features)

# =========================================================
# 3. PREPROCESSOR
# =========================================================
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)

# =========================================================
# 4. MODEL SELECTION FUNCTION
# =========================================================
def select_best_model(name, model_configs, X_train, y_train, X_test, y_test):
    best_result = None
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

    # keep input with feature names
    X_train_df = pd.DataFrame(X_train, columns=X_train.columns)
    X_test_df = pd.DataFrame(X_test, columns=X_test.columns)
    y_train_sr = pd.Series(y_train)
    y_test_sr = pd.Series(y_test)

    for cfg in model_configs:
        model = cfg["model"]

        model.fit(X_train_df, y_train_sr)

        train_pred = model.predict(X_train_df)
        test_pred = model.predict(X_test_df)

        train_acc = accuracy_score(y_train_sr, train_pred)
        test_acc = accuracy_score(y_test_sr, test_pred)
        gap = abs(train_acc - test_acc)

        cv_scores = cross_val_score(
            model,
            X_train_df,
            y_train_sr,
            cv=cv,
            scoring="accuracy",
            n_jobs=-1
        )
        cv_mean = cv_scores.mean()

        score = (0.55 * test_acc) + (0.35 * cv_mean) - (0.60 * gap)

        result = {
            "name": name,
            "model": model,
            "train_acc": train_acc,
            "test_acc": test_acc,
            "gap": gap,
            "cv_mean": cv_mean,
            "score": score,
            "params": cfg["params"]
        }

        if best_result is None or result["score"] > best_result["score"]:
            best_result = result

    print("\n" + "#" * 70)
    print(f"Best {name} parameters")
    print("#" * 70)
    print(best_result["params"])
    print(f"CV Accuracy   : {best_result['cv_mean']:.4f}")
    print(f"Train Accuracy: {best_result['train_acc']:.4f}")
    print(f"Test Accuracy : {best_result['test_acc']:.4f}")
    print(f"Gap           : {best_result['gap']:.4f}")

    return best_result

# =========================================================
# 5. EVALUATION FUNCTION
# =========================================================
def evaluate_model(name, model, X_train, y_train, X_test, y_test):
    X_train_df = pd.DataFrame(X_train, columns=X_train.columns)
    X_test_df = pd.DataFrame(X_test, columns=X_test.columns)
    y_train_sr = pd.Series(y_train)
    y_test_sr = pd.Series(y_test)

    model.fit(X_train_df, y_train_sr)

    y_train_pred = model.predict(X_train_df)
    y_test_pred = model.predict(X_test_df)

    train_acc = accuracy_score(y_train_sr, y_train_pred)
    test_acc = accuracy_score(y_test_sr, y_test_pred)
    gap = abs(train_acc - test_acc)

    print("\n" + "=" * 70)
    print(f"{name} Evaluation")
    print("=" * 70)
    print(f"Train Accuracy: {train_acc:.4f}")
    print(f"Test Accuracy : {test_acc:.4f}")
    print(f"Gap           : {gap:.4f}")

    print("\nTrain Classification Report")
    print(classification_report(y_train_sr, y_train_pred))

    print("Test Classification Report")
    print(classification_report(y_test_sr, y_test_pred))

    return {
        "Model": name,
        "Train Accuracy": train_acc,
        "Test Accuracy": test_acc,
        "Gap": gap
    }
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score

# =========================================================
# CALCULATE ACCURACY FOR EACH MODEL
# =========================================================
accuracy_results = []

models = [
    ("Logistic Regression", best_lr["model"]),
    ("Random Forest", best_rf["model"]),
    ("XGBoost", best_xgb["model"]),
    ("LightGBM", best_lgb["model"])
]

for name, model in models:

    # Predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    # Accuracy
    train_acc = accuracy_score(y_train, y_train_pred)
    test_acc = accuracy_score(y_test, y_test_pred)

    accuracy_results.append({
        "Model": name,
        "Train Accuracy": train_acc,
        "Test Accuracy": test_acc,
        "Gap": abs(train_acc - test_acc)
    })

# =========================================================
# CREATE TABLE
# =========================================================
accuracy_df = pd.DataFrame(accuracy_results)

# Sort by best balance + accuracy
accuracy_df = accuracy_df.sort_values(
    by=["Gap", "Test Accuracy"],
    ascending=[True, False]
)

print("\nAccuracy Comparison Table:")
print(accuracy_df.to_string(index=False))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

sns.set_style("white")

def plot_confusion_matrix_style(name, model, X_test, y_test, ax):
    y_pred = model.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)

    sns.heatmap(
        cm,
        annot=True,
        fmt="d",
        cmap="viridis",
        cbar=True,
        square=True,
        linewidths=0.5,
        linecolor="lightgray",
        annot_kws={"color": "red", "size": 14},
        ax=ax
    )

    ax.set_title(name, fontsize=16, fontweight="bold")
    ax.set_xlabel("Predicted label", fontsize=14, fontweight="bold")
    ax.set_ylabel("True label", fontsize=14, fontweight="bold")
    ax.set_xticklabels(["0", "1"], fontsize=12)
    ax.set_yticklabels(["0", "1"], fontsize=12, rotation=0)

# 2x2 layout
fig, axes = plt.subplots(2, 2, figsize=(14, 12))

models = [
    ("Logistic Regression", best_lr["model"]),
    ("Random Forest", best_rf["model"]),
    ("XGBoost", best_xgb["model"]),
    ("LightGBM", best_lgb["model"])
]

for (name, model), ax in zip(models, axes.flatten()):
    plot_confusion_matrix_style(name, model, X_test, y_test, ax)

plt.tight_layout()
plt.show()
```
